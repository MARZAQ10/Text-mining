```python
import pandas as pd
from tqdm.notebook import tqdm
import snscrape.modules.twitter as sntwitter
```

```python
scraper = sntwitter.TwitterSearchScraper("#palestine")
```

```python
tweets = []
desired_language = "en"
for i,tweet in enumerate(scraper.get_items()):
  if tweet.lang == desired_language:  
    data = [
        tweet.date,
        tweet.id,
        tweet.content,
        tweet.user.username,
        tweet.likeCount,
        tweet.retweetCount,
    ]
    tweets.append(data)
    if i>10000:
       break
```

```python
tweet_df = pd.DataFrame(
    tweets,columns = ["date","id","content","username","like count","retweet count"])
```
